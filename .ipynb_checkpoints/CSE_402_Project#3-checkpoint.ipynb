```python
def LBP_function(pixel_container):
    LBP_mtx = [[0 for j in range(100)] for i in range(100)]
    for i in range(1, 99):
        for j in range(1, 99):
            binary_string = ''
            tp_idx = tuple([i, j])
            first = i
            last = j - 1
            while tuple([first, last]) != tp_idx:
                if first == i and last == j - 1:
                    if pixel_container[i][j] < pixel_container[first][last]:
                        binary_string += str(1)
                        first += 1
                    else:
                        binary_string += str(0)
                        first += 1
                elif first == i + 1 and last == j - 1:
                    if pixel_container[i][j] < pixel_container[first][last]:
                        binary_string += str(1)
                        last += 1
                    else:
                        binary_string += str(0)
                        last += 1
                elif first == i + 1 and last == j:
                    if pixel_container[i][j] < pixel_container[first][last]:
                        binary_string += str(1)
                        last += 1
                    else:
                        binary_string += str(0)
                        last += 1
                elif first == i + 1 and last == j+1:
                    if pixel_container[i][j] < pixel_container[first][last]:
                        binary_string += str(1)
                        first -= 1
                    else:
                        binary_string += str(0)
                        first -= 1
                elif first == i and last == j+1:
                    if pixel_container[i][j] < pixel_container[first][last]:
                        binary_string += str(1)
                        first -= 1
                    else:
                        binary_string += str(0)
                        first -= 1
                elif first == i-1 and last == j+1:
                    if pixel_container[i][j] < pixel_container[first][last]:
                        binary_string += str(1)
                        last -= 1
                    else:
                        binary_string += str(0)
                        last -= 1
                elif first == i-1 and last == j:
                    if pixel_container[i][j] < pixel_container[first][last]:
                        binary_string += str(1)
                        last -= 1
                    else:
                        binary_string += str(0)
                        last -= 1
                else:
                    if pixel_container[i][j] < pixel_container[first][last]:
                        binary_string += str(1)
                        first += 1
                        last += 1
                    else:
                        binary_string += str(0)
                        first += 1
                        last += 1

            gray_scale_val = int(binary_string, 2)
            LBP_mtx[i][j] = gray_scale_val
            
    return LBP_mtx
     
```

```python
L0_image = LBP_function(pixel_mod_lst)
plt.imshow(L0_image, cmap = 'gray')
```

```python
def mean_face(face_lst):
    a = []
    for file in face_lst:
        img = PIL.Image.open(file)
        image_sequence = img.getdata()
        image_array = np.array(image_sequence)
        a.append(image_array)
    vector = np.sum(a, axis=0) 
    mean_vector = vector/30
    mean_face = mean_vector.reshape(30, 30)
    return mean_face
```

```python
def eigencoeff_function(face_lst2, other_list, matrix):
    df = pd.DataFrame()
    for file2 in face_lst2:
        img2 = PIL.Image.open(file2)
        image_sequence2 = img2.getdata()
        image_array2 = np.array(image_sequence2)
        z = image_array2 - other_list
        e = np.dot(matrix, z)
        df = df.append(pd.DataFrame(e))
        pixel_lst2 = []
        
    
    s = np.array(df)
    return np.reshape(s, (50, 50))
```
